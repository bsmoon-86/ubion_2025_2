```python
# 라이브러리 로드 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# 라인 그래프 생성 
# plot() 함수를 이용
# 인자가 1개면 y축에 데이터가 대입 
plt.plot( [10, 3, 5, 1] )
# show() -> print() 함수와 흡사한 기능
plt.show()
```

```python
# plot() 함수에 인자의 개수가 2개인 경우 
# 첫번째 인자는 x축 , 두번째 인자는 y축 대입
plt.plot(
    [1, 3, 7, 15], 
    [10, 3, 5, 1]
)
plt.show()
```

```python
# 선의 색상이나 종류를 변경 
plt.plot(
    [1,2,3,4], 
    [10, 5, 3, 1], 
    '--r'
)
plt.show()
```

```python
# 선의 색상을 헥스코드로 변경 
plt.plot(
    [1,2,3,4], 
    [5, 7, 8, 10], 
    "#259021"
)
plt.show()
```

```python
# x축, y축 그래프에 이름을 붙이는 방법
plt.plot(
    [1,2,3,4], 
    [10, 20, 30, 40]
)
plt.xlabel('name')
plt.ylabel('score')
plt.title('graph')
plt.show()
```

```python
# x, y 이름을 한글로 지정하는 경우 
# 한글이 깨짐 현상 발생 ( matplot에서 기본적으로 사용하는 폰트가 한글 지원X )
# 해당 문제는 해결하려면?
# matplot의 폰트 설정을 변경 
plt.plot(
    [1,2,3,4], 
    [10, 20, 30, 40]
)
plt.xlabel('이름')
plt.ylabel('점수')
plt.show()
```

```python
# 범례 추가 
x_list = [1,2,3,4]
y_list = [10, 20, 30, 40]
y2_list = [40, 30, 20, 10]
plt.plot(x_list, y_list)
plt.plot(x_list, y2_list)

# 범례 추가 (위치는 베스트 위치 선을 가리지 않은 최적의 위치로 선정)
# 위치는 고정하려면 loc 매개변수에 위치를 지정 
plt.legend(['A', 'B'], loc='upper center')

plt.show()
```

```python
# 막대 그래프 
plt.bar(x_list, y_list)

plt.show()
```

```python
# 수평으로 이루어진 막대그래프 
plt.barh(x_list, y_list)

plt.show()
```

```python
# 막대의 색상을 변경 너비 변경 
plt.bar(x_list, y_list, color=['r', 'g', 'b'], width=0.2)

plt.show()
```

```python
# 특정 막대의 색상을 변경 
graph = plt.bar(x_list, y_list)
# 3번째 막대를 선택(index는 2)
graph[2].set_color('g')

plt.show()
```

```python
# 데이터 로드 
drinks = pd.read_csv("../../csv/drinks.csv")
```

```python
drinks.head()
```

```python
plt.scatter(
    drinks['beer_servings'], 
    drinks['spirit_servings']
)
plt.show()
```

```python
# pandas 라이브러리 안에 있는 상관계수를 출력해주는 함수 
# corr()
# method 매개변수 (상관계수를 구하는 방법 설정)
# drinks 안에 beer_servings 부터 alcohol 컬럼까지 상관계수 확인

# drinks.drop(['country', 'continent'], axis=1)
# drinks[ ['beer_servings', 'spirit_servings', 'wine_servings', 
#          'total_litres_of_pure_alcohol'] ]
# drinks.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol' ]
# drinks.iloc[ : , 1:5]
drinks_corr = drinks.iloc[ : , 1 : -1].corr(method='pearson')
```

```python
# 히트맵 출력 
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    drinks_corr.values, 
    xticklabels= cols_view, 
    yticklabels= cols_view, 
    annot = True,   # 히트맵에 숫자를 표기할것인가?
    annot_kws= {'size' : 20}, # 숫자의 크기를 조절
    fmt = '0.3f',   # 숫자의 소수점 자리수 지정
    cmap= 'Greens',     #히트맵에 기준이 되는 색상을 지정
    square= False,       # 히트맵의 사각형을 정사각형으로 표시할것인가?
    cbar = True        # 사이드바의 존재 유무
)

plt.show()
```

```python
# 데이터의 분포도를 확인 
sns.pairplot(
    drinks.iloc[ : , 1:-1], 
    corner= True
)
plt.show()
```
