```python
# 라이브러리 로드 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# 라인 그래프 생성 
# plot() 함수를 이용
# 인자가 1개면 y축에 데이터가 대입 
plt.plot( [10, 3, 5, 1] )
# show() -> print() 함수와 흡사한 기능
plt.show()
```

```python
# plot() 함수에 인자의 개수가 2개인 경우 
# 첫번째 인자는 x축 , 두번째 인자는 y축 대입
plt.plot(
    [1, 3, 7, 15], 
    [10, 3, 5, 1]
)
plt.show()
```

```python
# 선의 색상이나 종류를 변경 
plt.plot(
    [1,2,3,4], 
    [10, 5, 3, 1], 
    '--r'
)
plt.show()
```

```python
# 선의 색상을 헥스코드로 변경 
plt.plot(
    [1,2,3,4], 
    [5, 7, 8, 10], 
    "#259021"
)
plt.show()
```

```python
# x축, y축 그래프에 이름을 붙이는 방법
plt.plot(
    [1,2,3,4], 
    [10, 20, 30, 40]
)
plt.xlabel('name')
plt.ylabel('score')
plt.title('graph')
plt.show()
```

```python
# x, y 이름을 한글로 지정하는 경우 
# 한글이 깨짐 현상 발생 ( matplot에서 기본적으로 사용하는 폰트가 한글 지원X )
# 해당 문제는 해결하려면?
# matplot의 폰트 설정을 변경 
plt.plot(
    [1,2,3,4], 
    [10, 20, 30, 40]
)
plt.xlabel('이름')
plt.ylabel('점수')
plt.show()
```

```python
# 내가 현재 사용하고 있는 컴퓨터에서 OS를 확인하기 위한 라이브러리리
import platform
```

```python
# 현재 컴퓨터의 OS을 출력
platform.system()
```

```python
# plt의 폰트 설정을 변경하는 방법법
# 만약에 MAC이라면?
if platform.system() == 'Darwin':
    plt.rc( 'font', family = 'AppleGothic' )
else:
    plt.rc('font', family = 'Malgun Gothic')
```

```python
# 범례 추가 
x_list = [1,2,3,4]
y_list = [10, 20, 30, 40]
y2_list = [40, 30, 20, 10]
plt.plot(x_list, y_list)
plt.plot(x_list, y2_list)

# 범례 추가 (위치는 베스트 위치 선을 가리지 않은 최적의 위치로 선정)
# 위치는 고정하려면 loc 매개변수에 위치를 지정 
plt.legend(['A', 'B'], loc='upper center')

plt.show()
```

```python
# 막대 그래프 
plt.bar(x_list, y_list)

plt.show()
```

```python
# 수평으로 이루어진 막대그래프 
plt.barh(x_list, y_list)

plt.show()
```

```python
# 막대의 색상을 변경 너비 변경 
plt.bar(x_list, y_list, color=['r', 'g', 'b'], width=0.2)

plt.show()
```

```python
# 특정 막대의 색상을 변경 
graph = plt.bar(x_list, y_list)
# 3번째 막대를 선택(index는 2)
graph[2].set_color('g')

plt.show()
```
