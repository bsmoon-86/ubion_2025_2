# 소수 나열하기 
- 어떤 정수 이하의 소수를 모두 나열하는 알고리즘을 구성
- 소수?
    - 자신 자신과 1 이외에는 나누어 떨어지지 않는 수를 의미 
        - 조건식 : 2부터 시작해서 n - 1 까지 어떤 정수로도 나누어 떨어지지 않는다. 

```python
# 2부터 1000까지의 정수중 소수는 무엇인가?

# 나눗셈의 횟수를 카운트를 하는 변수를 생성 
counter = 0
# 소수의 데이터를 담아둘 빈 리스트를 생성 
prime_list = []

# 2부터 1000까지 반복을 하는 반복문을 생성
for n in range(2, 1001):
    # 2부터 시작해서 n-1까지 반복 실행하는 반복문을 생성 
    for i in range(2, n):
        # 나눗셈의 카운트를 1씩 증가 
        counter += 1
        # n과 i를 나눈 나머지 값이 0인 경우 -> 소수가 아니다.
        if n % i == 0:
            # 2번째 반복문을 강제 종료
            break
    # else는 for문에서 if 조건식이 모두 거짓인 경우에만 실행
    else:
        prime_list.append(n)

print(f"총 나눗셈의 횟수는 {counter}")
print(f"2부터 1000까지의 소수의 배열은 {prime_list}")
```

```python
# 소수나열하기 개선 
# 소수는 짝수는 존재하지 않는다. 

counter = 0
prime_list = []

# 2는 소수임으로 prime_list에 추가
prime_list.append(2)

# 3부터 1000까지의 홀수 값들을 반복 실행
for n in range(3, 1001, 2):
    for i in range(2, n):
        counter += 1
        if n % i == 0 :
            break
    else:
        prime_list.append(n)

print(f"총 나눗셈의 횟수는 {counter}")
print(f"2부터 1000까지의 소수의 배열은 {prime_list}")
```

```python
# 개선 
# 홀수 데이터를 기준으로 반복 -> 이미 찾은 소수로 나눠서 나머지가 0이라면 소수가 아니다. 

counter = 0 
prime_list = [2]

for n in range(3, 1001, 2):
    # 이미 찾은 소수를 하나씩 추출
    for i in range(1, len(prime_list)):
        counter += 1
        if n % prime_list[i] == 0:
            break
    else:
        prime_list.append(n)


print(f"총 나눗셈의 횟수는 {counter}")
print(f"2부터 1000까지의 소수의 배열은 {prime_list}")
```

```python
# 개선 
# n의 제곱근 이하의 어떤 소수로도 나누어 떨어지지 않는다. 
# 조건식 -> prime_list[i] * prime_list[i] <= n인 경우에는 소수 데이터로 n을 나눠서 확인

counter = 0

prime_list = [2,3]

for n in range(5, 1001, 2):
    i = 1
    while prime_list[i] * prime_list[i] <= n:
        counter += 2
        if n % prime_list[i] == 0:
            break
        # while의 조건식이 거짓이 될수 있도록 값을 변경 
        i += 1
    else:
        prime_list.append(n)
        counter += 1


print(f"총 나눗셈의 횟수는 {counter}")
print(f"2부터 1000까지의 소수의 배열은 {prime_list}") 
```
